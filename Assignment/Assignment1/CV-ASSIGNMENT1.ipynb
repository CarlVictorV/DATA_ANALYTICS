```python
# Check the distribution of each column and determine which probability distribution it fits.
import matplotlib.pyplot as plt

# Create a figure and axis
fig, ax = plt.subplots(4, 4, figsize=(15, 15))

# Flatten the axis array for easier iteration
ax_flat = ax.flatten()

# Loop over each column
for i, column in enumerate(X.columns):
    # Create a histogram
    ax_flat[i].hist(X[column], bins=20, edgecolor='black')
    ax_flat[i].set_title(column)

# Hide any empty subplots
for i in range(len(X.columns), len(ax_flat)):
    ax_flat[i].axis('off')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

for column in X.columns:
    # Remove non-finite values
    data = pd.to_numeric(X[column], errors='coerce')
    data = data[np.isfinite(data)]
    
    # Check if data is not empty
    if not data.empty:
        # Fit the data to different distributions
        dist_names = ['gamma', 'beta', 'norm', 'lognorm']
        best_fit_name = None
        best_kstest = 100  # Initialize with a large value

        for dist_name in dist_names:
            # Fit the data to the distribution
            dist = getattr(stats, dist_name)
            params = dist.fit(data)
            
            # Perform the Kolmogorov-Smirnov test
            _, p = stats.kstest(data, dist_name, args=params)
            
            # Update the best fit distribution if the p-value is better
            if p < best_kstest:
                best_kstest = p
                best_fit_name = dist_name

        print(f"Column '{column}' best fits {best_fit_name} distribution with p-value {best_kstest}")
    else:
        print(f"Column '{column}' has no valid data points.")
```

```python
# Select only numeric columns
X_numeric = X.select_dtypes(include=[np.number])

# Create a summary statistics table
summary = X_numeric.describe().T

# Add the median
summary['median'] = X_numeric.median()

# Add the skew
summary['skew'] = X_numeric.skew()

# Add the kurtosis
summary['kurtosis'] = X_numeric.kurtosis()

# Display the summary
print(summary)
```
