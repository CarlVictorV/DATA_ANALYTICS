### Data: Siegler,R.. (1994). Balance Scale. UCI Machine Learning Repository. https://doi.org/10.24432/C5488X.

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
balance_scale = fetch_ucirepo(id=12) 
  
# data (as pandas dataframes) 
X = balance_scale.data.features 
y = balance_scale.data.targets 
  
# metadata 
print(balance_scale.metadata) 
  
# variable information 
print(balance_scale.variables) 
```

# DO DECISION TREE

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()
y = labelencoder_X.fit_transform(y).ravel()

# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)

# Fitting Decision Tree Classification to the Training set
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)

print("Confusion Matrix: ")
print(cm)
print("Accuracy: ")
print(accuracy)


#Visualizing the Decision Tree
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20))
tree.plot_tree(classifier, filled=True, feature_names=X.columns.tolist(), class_names=['L', 'B', 'R'])
plt.show()
```
